````polyglot-notebook
[Description("Provides a random vacation destination.")]
static string GetRandomDestination()
{
    var destinations = new List<string>
    {
        "Paris, France",
        "Tokyo, Japan",
        "New York City, USA",
        "Sydney, Australia",
        "Rome, Italy",
        "Barcelona, Spain",
        "Cape Town, South Africa",
        "Rio de Janeiro, Brazil",
        "Bangkok, Thailand",
        "Vancouver, Canada"
    };

    var random = new Random();
    int index = random.Next(destinations.Count);
    return destinations[index];
}
````

## üé≤ Define Tool Function: Random Destination Generator

Create a tool function that the agent can call to get random vacation destinations.
The `[Description]` attribute enables the AI to understand when and how to use this tool.

````polyglot-notebook
var github_endpoint = Environment.GetEnvironmentVariable("GITHUB_ENDPOINT") ?? throw new InvalidOperationException("GITHUB_ENDPOINT is not set.");
var github_model_id = Environment.GetEnvironmentVariable("GITHUB_MODEL_ID") ?? "gpt-4o-mini";
var github_token = Environment.GetEnvironmentVariable("GITHUB_TOKEN") ?? throw new InvalidOperationException("GITHUB_TOKEN is not set.");
````

## üîë Load Configuration from Environment Variables

Extract and validate the required configuration settings:
- **GITHUB_ENDPOINT**: API endpoint URL for GitHub Models
- **GITHUB_MODEL_ID**: The AI model identifier (e.g., gpt-4o-mini)
- **GITHUB_TOKEN**: Authentication token for API access

````polyglot-notebook
var openAIOptions = new OpenAIClientOptions()
{
    Endpoint= new Uri(github_endpoint)
};
````

## ‚öôÔ∏è Configure OpenAI Client Options

Set up the OpenAI client configuration to redirect API calls to GitHub Models endpoint.

````polyglot-notebook

var openAIClient = new OpenAIClient(new ApiKeyCredential(github_token), openAIOptions);
````

## üîå Initialize OpenAI Client

Create the OpenAI client instance configured for GitHub Models integration.

````polyglot-notebook
AIAgent agent = new OpenAIClient(new ApiKeyCredential(github_token), openAIOptions).GetChatClient(github_model_id).AsIChatClient().CreateAIAgent(
    instructions:"You are a helpful AI Agent that can help plan vacations for customers at random destinations", tools: [AIFunctionFactory.Create((Func<string>)GetRandomDestination)]);
````

## ü§ñ Create AI Agent with Travel Planning Capabilities

Build the complete agent pipeline: OpenAI Client ‚Üí Chat Client ‚Üí AI Agent.
Configure the agent with travel planning instructions and register the destination tool.

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

## üßµ Create Conversation Thread

Initialize a new conversation thread for maintaining context across multiple agent interactions.

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Plan me a day trip",thread));
````

## üöÄ Execute First Travel Planning Request

Run the agent with an initial travel planning request. The agent will analyze the request, use the tool if needed, and generate an itinerary.

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("I don't like that destination. Plan me another vacation.",thread));
````
